```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.sin(x), np.cos(x)
ax.plot(x, s, label = "Sine", color = "blue")
ax.plot(x, c, label = "Cosine", color = "red")
ax.set(title="Trigonometry", xlabel="Angles", ylabel="Function", 
       xticks=np.linspace(0, 2*np.pi, 5), yticks=np.linspace(-1, 1, 9))
ax.legend()
ax.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate Sample Data (5Hz Sine Wave + Noise)
t = np.linspace(0, 1, 100)
signal = np.sin(2 * np.pi * 5 * t) + 0.2 * np.random.randn(100)
categories = ['A', 'B', 'C']
values = [5, 8, 3]

# 2. Create a 2x3 Grid
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

# --- [0,0] Line Plot (Time Domain) ---
ax[0, 0].plot(t, signal, color='blue', label='ECG')
ax[0, 0].set(title="Line: Continuous Signal", xlabel="s", ylabel="V")

# --- [0,1] Scatter Plot (Samples) ---
ax[0, 1].scatter(t, signal, color='red', alpha=0.5)
ax[0, 1].set(title="Scatter: Data Points")

# --- [0,2] Histogram (Noise) ---
ax[0, 2].hist(signal, bins=15, color='green', edgecolor='black')
ax[0, 2].set(title="Histogram: Distribution")

# --- [1,0] Bar Chart (Comparison) ---
ax[1, 0].bar(categories, values, color='purple')
ax[1, 0].set(title="Bar: Category Comparison")

# --- [1,1] Box Plot (Outliers) ---
ax[1, 1].boxplot(signal, patch_artist=True, boxprops=dict(facecolor='pink'))
ax[1, 1].set(title="Box: IQR & Outliers", xticklabels=['Signal'])

# --- [1,2] Stem Plot (Discrete) ---
ax[1, 2].stem(t[::5], signal[::5]) # Subsampled for clarity
ax[1, 2].set(title="Stem: Sampled Signal")

# 3. Global Polish
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Normal Sine', 'Noise/Interference', 'Tachycardia', 'Bradycardia']
sizes = [70, 15, 10, 5]  # Percentages or raw values
colors = ['#66b3ff', '#ff9999', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0, 0)  # "explode" the 1st slice (Normal Sine)

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Plotting the pie chart
ax.pie(sizes, 
       explode=explode, 
       labels=labels, 
       colors=colors, 
       autopct='%1.1f%%', 
       shadow=True, 
       startangle=140)

# Set the title
ax.set_title("ECG Signal Classification Proportions")

plt.show()
```
