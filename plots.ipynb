```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.sin(x), np.cos(x)
ax.plot(x, s, label = "Sine", color = "blue")
ax.plot(x, c, label = "Cosine", color = "red")
ax.set_ylabel("Trigonometric function")
ax.set_xlabel("Angles")
ax.set_xticks(np.linspace(0, 2*np.pi, 5))
ax.set_yticks(np.linspace(-1, 1, 9))
ax.legend()
ax.grid(True)
plt.show()
```
