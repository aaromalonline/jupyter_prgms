```python
import os
os.environ["QT_QPA_PLATFORM"] = "xcb" # Fixes the Wayland/Qt error
import cv2

img = cv2.imread('./datasets/lm.jpg', cv2.IMREAD_COLOR)

# Correct usage: Provide a window name AND the image
cv2.imshow('Original Image', img) 

new_width, new_height = 300, 300
resized_img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

cv2.imshow('Resized Image', resized_img)

cv2.waitKey(0) # Wait for you to press a key
cv2.destroyAllWindows() # Clean up the windows properly
```
